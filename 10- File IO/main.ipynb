# File IO (Input / Output)

- open : این متود میاد فایل مورد نظر مارو با میکنه و با استفاده از متود هایی که داره به ما اجازه میده که داخل فایلو بخونیم یا بنویسیم روش و کار های مختلف دیگه

- .seek() : این متود کارش اینه که مارو به هر جای خط از فایل بخوایم بر میگردونه. اگر فایلو رید بکنیم به همون جایی که سیک رو فراخانی کردیم برمیگرده.
- .readline() : با استفاده از این متود ما خط به خط فایلمونو میتونیم بخونیم هربار که کال بشه فقط یک خطو میخونه. به ترتیب ادامه میده.
- .readlines() : تمامی خط های مارو در قالب یک لیست در میاره
- .close() : این میاد فایل مارو میبنده و جایی که نیاز نداشته باشیم به کار کردن با فایل در کدمون میایم فایلو میبنیدم تا مصرف منابع کمتری داشته باشیم و مموری کمتری اشغال بشه.
- with : این استیتمنت کارش اینه که ما جای اینکه بیایم به صورت دستی از متود کلوز استفاده بکنیم میایم از این استفاده میکنیم و این به ما این اعتمادو میده که بعد از انجام عملیات فایل مارو میبنده که این استیتمنت توی مدیریت منابع و سرعت تاثیر به سزایی داره

- file mode : 
    - r : فایلو میشه خوند
    - r+ : هم میشه فایلو خوند هم روش نوشت
    - a : فایلو مینویسی روش با این تفاوت که به اخر فایل ما نوشته میشه.
    - w : این مود یعنی نوشتن روی فایل ما میتونیم با این مود بنویسیم ولی با این تفاوت که میاد تمامی محتواهای قبلی فایل مارو حذف میکنه و نوشته جدید مارو وارد فایل میکنه. اگر فایل مورد نظر ما وجود نداشته باشه این مود برای ما فایل جدید میسازه.
